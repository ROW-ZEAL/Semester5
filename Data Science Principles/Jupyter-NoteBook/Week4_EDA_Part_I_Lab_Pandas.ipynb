### Question 1. Calculate the average marks of each student (make sure to insert a new column name “Average Marks” in the dataframe.) (Creating a new column)
To create a new column, use the [] brackets with the new column name at the left side of the assignment.

```python
# Add new Column with 0.0 value
# df["Average_Marks"] = 0.0
```

```python
# Add mean value of this 3's to average marks
df["Average_Marks"] = df[["DSP_Marks", "SSM_Marks", "AP_Marks"]].mean(axis = 1)

# axis = 1 (rows (horizentally)) - By default axis = 0 (column)
# By default, NaN value ignore while calculating mean
```

### Question 2. Count the number of students who have got above 80 average marks. (Select a subset)

```python
avg_marks = df[df["Average_Marks"] >= 80].shape[0]

print(f"No. of Student who have got above 80 marks: {avg_marks}")
```
